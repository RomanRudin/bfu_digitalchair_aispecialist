### 19. Создайте pivot_table по любым столбцам и опишите полученный результат. 

Средняя температура в рабочие и выходные дни в зависимости от времени года.

```python
pivot_t = new_df.pivot_table('temp', 'workingday' , 'name_season', aggfunc='mean')
pivot_t
```

### 20. Получите процентное соотношение count в зависимости от времен года. 

```python
pivot_t = new_df.pivot_table('count', 'name_season', aggfunc='sum')
pivot_t.div(pivot_t.sum(0))*100
```
