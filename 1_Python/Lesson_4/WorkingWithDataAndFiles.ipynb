### Чтение содержимого файла в список

```python
from os import getcwd
with open(fr"{getcwd()}\data\user_ids.txt", 'r', encoding='utf-8') as file:
    user_id = list(map(int, file.read().splitlines()))
print("Первые пять элементов: ", ', '.join(list(map(str, user_id[:5]))))
print("Последний из набора: ", user_id[4])
print("Количество элементов: ", len(user_id))
print("Количество уникальных значений: ", len(set(user_id)))
```
