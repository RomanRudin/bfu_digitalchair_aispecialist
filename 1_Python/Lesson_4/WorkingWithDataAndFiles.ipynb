### Чтение содержимого файла в список

### Перевод значений в численный вид

```python
from os import getcwd
with open(fr"{getcwd()}\data\data_3_columns.txt", 'r', encoding='utf-8') as file:
    for line in file: 
        line = line.strip().split('\t') 
        medium = line[0] 
        source = line[1] 
        amount_paid = float(line[2].replace(',', '.'))
        print(line) 
        print(source, medium, amount_paid)
```

### Подсчет суммы в файлах

```python
from os import getcwd
total_sum = 0
with open(fr"{getcwd()}\data\data_3_columns.txt", 'r', encoding='utf-8') as file:
    for line in file: 
        line = line.strip().split('\t')
        amount_paid = float(line[2].replace(',', '.'))
        total_sum += amount_paid
        print('Текущая сумма расходов: {:.2f}'.format(total_sum)) 
```

```python
print('Сумма расходов в последнем примере: {:.2f}'.format(total_sum)) 
```

```python
from os import getcwd
total_sum = 0
with open(fr"{getcwd()}\data\data_3_columns.txt", 'r', encoding='utf-8') as file:
    for line in file: 
        line = line.strip().split('\t')
        medium = line[0] 
        source = line[1]
        amount_paid = float(line[2].replace(',', '.'))
        if source == 'google':
            total_sum += amount_paid
            print('Текущая сумма расходов google: {:.2f}'.format(total_sum))
```

```python
from os import getcwd
total_sum = 0
with open(fr"{getcwd()}\data\data_3_columns.txt", 'r', encoding='utf-8') as file:
    for line in file: 
        line = line.strip().split('\t')
        medium = line[0] 
        source = line[1]
        amount_paid = float(line[2].replace(',', '.'))
        if (source == 'yandex') and (medium == "seo"):
            total_sum += amount_paid
print("Сумма для строк, у которых source == 'yandex' и medium == 'seo': {:.2f}".format(total_sum))
```
